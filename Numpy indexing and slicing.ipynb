### Broadcasting
Numpy arrays differ from a normal Python list because of their ability to broadcast:

```python
arr[0:5] = 100
```

```python
arr
```

```python
arr = np.arange(0,11)
```

```python
arr
```

```python
slice_of_arr = arr[0:6]
```

```python
slice_of_arr
```

```python
slice_of_arr[:] =99
```

```python
slice_of_arr
```

```python
arr
```

```python
arr_copy = arr.copy()
```

```python
arr
```

```python
arr_copy
```

```python
arr_copy[:] =100
```

```python
arr_copy
```

```python
arr
```

### Selection
Let's briefly go over how to use brackets for selection based off of comparison operators.

```python
arr = np.arange(1,11)
arr
```

```python
arr > 4
```

```python
bool_arr = arr > 4
```

```python
bool_arr
```

```python
arr[bool_arr]
```

```python
arr[arr>4]
```

```python
arr[arr<3]
```

```python
arr_2d =np.arange(50).reshape(5,10)
```

```python
arr_2d
```

```python
arr_2d[1:3,3:5]
```

```python
arr_2d[2:4,3:5]
```
